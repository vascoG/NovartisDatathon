```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
PATH_TRAIN = Path('../data/train_data.parquet')
```

```python
train_data = pd.read_parquet(PATH_TRAIN)

train_data.head(n=35)
```

```python
print(train_data.duplicated(subset=['brand', 'country', 'date']).sum())

print(train_data.info())
```

## Correlation and Skewness

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = train_data.corr()
sns.heatmap(corr,
    cmap=sns.diverging_palette(220, 10, as_cmap=True),
    vmin=-1.0, vmax=1.0,
    square=True, ax=ax)
```

```python
train_columns = ['brand', 'country', 'date', 'phase', 'dayweek', 'month', 'wd', 'hospital_rate', 'ther_area', 'main_channel', 'monthly', 'n_nwd_bef', 'n_nwd_aft']

train_data = train_data[train_columns]

train_data.head(n=5)
```

```python
train_data.hist(figsize=(20, 20))
```

```python
# categorize features

train_data['brand'] = train_data['brand'].astype('category')
train_data['country'] = train_data['country'].astype('category')
train_data['dayweek'] = train_data['dayweek'].astype('category')
train_data['month'] = train_data['month'].astype('category')
train_data['wd'] = train_data['wd'].astype('category')
```

```python
print(train_data.info())
```

```python
for col in ['ther_area', 'main_channel','brand','country','dayweek','month','wd']:
    sns.countplot(x=col, data=train_data)
    plt.show()
```

```python
for col in ['ther_area', 'main_channel','brand','country','dayweek','month','wd']:
    sns.boxplot(x=col, y='phase', data=train_data)
    plt.show()
```
